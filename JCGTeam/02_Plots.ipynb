```python
import pickle
import matplotlib.pyplot as plt
```

# Con1D from Lab

```python
# Load the history of the model
with open('./Gianmario/SavedModelsForPlotting/Conv1D_BASE/history.pickle', 'rb') as file_pi:
   history_Conv1D_BASE = pickle.load(file_pi)
   
with open('./Gianmario/SavedModelsForPlotting/Conv1D_MIDDLE/history.pickle', 'rb') as file_pi:
   history_Conv1D_MIDDLE = pickle.load(file_pi)

with open('./Gianmario/SavedModelsForPlotting/Conv1D_BIG/history.pickle', 'rb') as file_pi:
   history_Conv1D_BIG = pickle.load(file_pi)
```

```python
# Plot the loss and accuracy of both models
metric = "sparse_categorical_accuracy"
plt.figure(figsize=(8, 3))
plt.subplot(1, 2, 1)
plt.plot(history_Conv1D_BASE[metric], label='Single Model (Training)', alpha=0.5, color='red')
plt.plot(history_Conv1D_BASE["val_" + metric], label='Single Model (Validation)', color='red')
plt.plot(history_Conv1D_MIDDLE[metric], label='Double Model (Training)', alpha=0.5, color='blue')
plt.plot(history_Conv1D_MIDDLE["val_" + metric], label='Double Model (Validation)', color='blue')
plt.plot(history_Conv1D_BIG[metric], label='Tripple Model (Training)', alpha=0.5, color='orange')
plt.plot(history_Conv1D_BIG["val_" + metric], label='Tripple Model (Validation)', color='orange')
plt.title('Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Loss')
#plt.legend()
plt.subplot(1, 2, 2)
metric = "loss"
plt.plot(history_Conv1D_BASE[metric], label='Single Model (Training)', alpha=0.5, color='red')
plt.plot(history_Conv1D_BASE["val_" + metric], label='Single Mode (Validation)l', color='red')
plt.plot(history_Conv1D_MIDDLE[metric], label='Double Model (Training)', alpha=0.5, color='blue')
plt.plot(history_Conv1D_MIDDLE["val_" + metric], label='Double Model (Validation)', color='blue')
plt.plot(history_Conv1D_BIG[metric], label='Tripple Model (Training)', alpha=0.5, color='orange')
plt.plot(history_Conv1D_BIG["val_" + metric], label='Tripple Model (Validation)', color='orange')
plt.title('Loss')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```
